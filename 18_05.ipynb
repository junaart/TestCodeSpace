```python
import pandas as pd

Data = pd.read_csv('https://raw.githubusercontent.com/junaart/ForStudents/refs/heads/main/Python/data_1.csv')
Data
```

```python
import numpy as np
```

```python
global  p
p = -1
h = widgets.Text(
    value='',
    description='ФИО:',
    disabled=False
)
b1 = widgets.Button(
    description='>',
    disabled=False
)
b2 = widgets.Button(
    description='<',
    disabled=False
)

b3 = widgets.Button(
    description='New_oklad',
    disabled=False
)

b4 = widgets.Button(
    description='New_add',
    disabled=False

d1 = widgets.DatePicker(
    description='Дата рождения: ',
    disabled=False
)

d2 = widgets.DatePicker(
    description='Дата приема на работу: ',
    disabled=False
)
s = widgets.RadioButtons(
    options=['м','ж'],
    description='Пол:',
    disabled=False
)

oklad = widgets.IntText(
    description='Оклад: ',
    value = 0
)

famely = widgets.Dropdown(
    options=[('женат', 1), ('разведен', 2), ('замужем', 3),['незамужем',4],['холост',5],['разведена',6]],
    value=1,
    description='Семейное положение:',
)

old_year = widgets.IntText(
    description='Возраст: ',
    value = 0
)

work_time = widgets.IntText(
    description='Стаж: ',
    value = 0
)

children= widgets.SelectMultiple(
    options=['0', '1', '2', '3', '4'],
    value=['0'],
    rows=5,
    description='Количество детей',
    disabled=False
)

oklad_new = widgets.IntText(
    description='Новый оклад: ',
    value = 0
)

def b1_clicked(bb):
  import datetime as dt
  global p
  p +=1
  if p >= len(Data.iloc[:,0]):
    p -= 1
  h.value = Data.iloc[p,0]
  dd = Data.iloc[p,1].split('.')
  d1.value = dt.date(int(dd[2]), int(dd[1]), int(dd[0]))
  dd = Data.iloc[p,3].split('.')
  d2.value = dt.date(int(dd[2]), int(dd[1]), int(dd[0]))
  s.value = Data.iloc[p,2]
  oklad.value = Data.iloc[p,4]
  if Data.iloc[p,5] == 'женат':
    famely.value = 1
  elif Data.iloc[p,5] == 'разведен':
    famely.value = 2
  elif Data.iloc[p,5] == 'замужем':
    famely.value = 3
  elif Data.iloc[p,5] == 'незамужем':
    famely.value = 4
  elif Data.iloc[p,5] == 'холост':
    famely.value = 5
  elif Data.iloc[p,5] == 'разведена':
    famely.value = 6
  else:
    famely.value = -1
  old_year.value = Data.iloc[p,8]
  work_time.value = Data.iloc[p,7]
  if np.isnan(Data.iloc[p,6]):
    children.value = ['0']
  else:
    children.value = [str(int(Data.iloc[p,6]))]



def b2_clicked(bb):
  import datetime as dt
  global p
  p -= 1
  if p < 0:
    p  = 0
  h.value = Data.iloc[p,0]
  dd = Data.iloc[p,1].split('.')
  d1.value =  dt.date(int(dd[2]), int(dd[1]), int(dd[0]))
  dd = Data.iloc[p,3].split('.')
  d2.value =  dt.date(int(dd[2]), int(dd[1]), int(dd[0]))
  s.value = Data.iloc[p,2]
  oklad.value = Data.iloc[p,4]
  if Data.iloc[p,5] == 'женат':
    famely.value = 1
  elif Data.iloc[p,5] == 'разведен':
    famely.value = 2
  elif Data.iloc[p,5] == 'замужем':
    famely.value = 3
  elif Data.iloc[p,5] == 'незамужем':
    famely.value = 4
  elif Data.iloc[p,5] == 'холост':
    famely.value = 5
  elif Data.iloc[p,5] == 'разведена':
    famely.value = 6
  else:
    famely.value = -1
  old_year.value = Data.iloc[p,8]
  work_time.value = Data.iloc[p,7]
  if np.isnan(Data.iloc[p,6]):
    children.value = ['0']
  else:
    children.value = [str(int(Data.iloc[p,6]))]

def b3_clicked(bb):
  global p
  Data.iloc[p,4] = oklad_new.value

def b4_clicked(bb):
  global p
  


b1.on_click(b1_clicked)
b2.on_click(b2_clicked)
b3.on_click(b3_clicked)
b4.on_click(b4_clicked)


display(h,d1,d2,s,oklad,famely,old_year,work_time,children,oklad_new,b1,b2,b3,b4)

```
